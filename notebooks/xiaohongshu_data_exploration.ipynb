```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['post_count'], ax=axes[0], kde=True)
axes[0].set_title('Post Count Distribution')
sns.histplot(df['avg_sentiment'], ax=axes[1], kde=True)
axes[1].set_title('Avg Sentiment Distribution')
plt.show()
```
